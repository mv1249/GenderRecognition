```python
import os
from PIL import Image
from glob import glob
```

```python
man_train = [x for x in glob('Train/Man/*')]
woman_train = [x for x in glob('Train/Woman/*')]
man_train
```

```python
os.mkdir('Training')
os.mkdir('Training/Man')
os.mkdir('Training/Woman')
os.mkdir('Testing')
os.mkdir('Testing/Man')
os.mkdir('Testing/Woman')
```

```python
man_train_list = []
woman_train_list = []
man_test_list = []
woman_test_list =[]
```

```python
man_train_re_list = []
woman_train_re_list = []
man_test_re_list = []
woman_test_re_list =[]
```

```python
for i in glob('Train/Man/*'):
    img = Image.open(i)
    man_train_list.append(img)

for image in man_train_list:
    image = image.resize((224,224))
    man_train_re_list.append(image)
    
for i,new in enumerate(man_train_re_list):
    new.save('{}{}{}'.format('Training/Man/',i+1,'.jpg'))
```

```python
for i in glob('Train/Woman/*'):
    img = Image.open(i)
    woman_train_list.append(img)

for image in woman_train_list:
    image = image.resize((224,224))
    woman_train_re_list.append(image)
    
for i,new in enumerate(woman_train_re_list):
    new.save('{}{}{}'.format('Training/Woman/',i+1,'.jpg'))
```

```python
for i in glob('Test/Man/*'):
    img = Image.open(i)
    man_test_list.append(img)

for image in man_test_list:
    image = image.resize((224,224))
    man_test_re_list.append(image)
    
for i,new in enumerate(man_test_re_list):
    new.save('{}{}{}'.format('Testing/Man/',i+1,'.jpg'))
```

```python
for i in glob('Test/Woman/*'):
    img = Image.open(i)
    woman_test_list.append(img)

for image in woman_test_list:
    image = image.resize((224,224))
    woman_test_re_list.append(image)
    
for i,new in enumerate(woman_test_re_list):
    new.save('{}{}{}'.format('Testing/Woman/',i+1,'.jpg'))
```
